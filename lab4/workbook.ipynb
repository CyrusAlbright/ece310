```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("lab4/scope_0.csv", skiprows=1)
t = data["second"].to_numpy() * 1000000
v_tri = data["Volt"].to_numpy()
v_squ = data["Volt.1"].to_numpy()

plt.plot(t, v_squ, 'g-', label="Square Wave")
plt.plot(t, v_tri, 'b-', label="Triangle Wave")

plt.xlabel("Time (us)")
plt.ylabel("Signal (V)")
plt.xlim(-25, 25)
plt.ylim(-0.5, 5.5)
plt.title("Oscillator Data")
plt.grid(True, axis="both")
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

tri_min, tri_max = np.min(v_tri), np.max(v_tri)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("lab4/scope_1.csv", skiprows=1).dropna()
t = data["second"].to_numpy() * 1000000
v_tri = data["Volt"].to_numpy()
v_squ = data["Volt.1"].to_numpy()

plt.plot(t, v_squ, 'g-', label="Square Wave")
plt.plot(t, v_tri, 'b-', label="Triangle Wave")

plt.xlabel("Time (us)")
plt.ylabel("Signal (V)")
plt.xlim(-25, 25)
plt.ylim(-0.5, 5.5)
plt.title("Oscillator Data")
plt.grid(True, axis="both")
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

tri_min, tri_max = np.min(v_tri), np.max(v_tri)
```

```python
fft_squ = np.abs(np.fft.fft(v_squ - np.mean(v_squ)))
fft_freq = np.fft.fftfreq(len(v_squ), d=(t[1]-t[0])*1e-6)
plt.plot(fft_freq, fft_squ)
plt.xlim(0, 6e5)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT of Square Wave")
plt.grid()
plt.show()

print(fft_freq[np.argmax(fft_squ)])
```
