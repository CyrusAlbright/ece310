```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

before = pd.read_csv("fft.csv", skiprows=1).dropna()
after = pd.read_csv("fft_low_pass.csv", skiprows=1).dropna()

f = before["Hertz"].to_numpy() / 1000
A_before = before["dBV"].to_numpy()
A_after = after["dBV"].to_numpy()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))
fig.suptitle("FFT Spectrum")
fig.supxlabel("Frequency (kHz)")
ax1.plot(f, A_before, color="blue")
ax1.set_title("Before Low-Pass Filter")
ax1.set_ylabel("Amplitude (dBV)")
ax1.set_xlim(0, 100)
ax1.set_ylim(-110, 10)
ax1.axvline(1, color="red", linestyle="--", label="Fundamental Frequency (1 kHz)")
ax1.axvline(96, color="black", linestyle="--", label="PWM Frequency (96 kHz)")
ax1.grid(True)
ax1.legend(loc="lower right")
ax2.plot(f, A_after, color="blue")
ax2.set_title("After Low-Pass Filter")
ax2.set_ylabel("Amplitude (dBV)")
ax2.set_xlim(0, 100)
ax2.set_ylim(-110, 10)
ax2.axvline(1, color="red", linestyle="--", label="Fundamental Frequency (1 kHz)")
ax2.axvline(96, color="black", linestyle="--", label="PWM Frequency (96 kHz)")
ax2.grid(True)
ax2.legend(loc="upper right")
fig.tight_layout()
```
