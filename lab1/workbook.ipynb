```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("scope_0.csv", skiprows=1)
t = data["second"].to_numpy()
v = data["Volt"].to_numpy()

plt.plot(t * 1000000, v)
plt.xlabel("Time (us)")
plt.ylabel("Signal (V)")
plt.title("Oscilloscope Probe Compensation Data")
plt.grid(True)
plt.show()
```
